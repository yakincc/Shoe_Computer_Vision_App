```python
import sqlite3
import pandas as pd
import numpy as np
import csv

conn = sqlite3.connect('database.db')
cursor = conn.cursor()

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
columns = ['id', 'brand', 'categories', 'colors', 'dateAdded', 'dateUpdated', 'descriptions', 'features', 'imageURLs', 'manufacturer', 'manufacturerNumber', 'name', 'prices.amountMin', 'prices.amountMax', 'prices.currency', 'prices.isSale', 'prices.offer', 'sizes', 'skus', 'sourceURLs']
men_data = pd.read_csv(r'https://query.data.world/s/44sfwttfnhuh6cqquxegcvz3jbnb5v?dws=00000', usecols = columns)
women_data = pd.read_csv(r'https://query.data.world/s/ldwkl7auyvfcuh4u5momo3z3yszphp?dws=00000', encoding='latin1', usecols = columns)
```

## Filter row with Null features

```python
clean_men_data = men_data.loc[men_data['features'].notnull()]
clean_men_data = clean_men_data.reset_index(drop = True)
clean_men_data.shape
```

```python
clean_women_data = women_data.loc[women_data['features'].notnull()]
clean_women_data = clean_women_data.reset_index(drop = True)
clean_women_data.shape
```

## Create db from existing dataframes

```python
clean_women_data.to_sql('women_shoe_data', conn, if_exists='replace', index=False)
clean_men_data.to_sql('men_shoe_data', conn, if_exists='replace', index=False)
```

```python
# Display the womens' database contents as a table using pandas
query = "SELECT * FROM women_shoe_data "
women_shoe_data = pd.read_sql_query(query, conn)
women_shoe_data
```

```python
# Display the mens' database contents as a table using pandas
query = "SELECT * FROM men_shoe_data "
men_shoe_data = pd.read_sql_query(query, conn)
men_shoe_data
```
